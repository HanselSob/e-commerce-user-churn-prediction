```python
# Import packages for data manipulation
import pandas as pd
import numpy as np
```

```python
# Load csv dataset into dataframe // https://www.kaggle.com/datasets/mehmetsabrikunt/internet-service-churn
df = pd.read_csv('internet_service_churn.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Isolate rows where churn is 1
churn_df = df[df['churn'] == 1]
# Display summary stats of rows with churn values
churn_df.describe()
```

```python
# Isolate rows where churn is 0
no_churn_df = df[df['churn'] == 0]
# Display summary stats of rows with churn values
no_churn_df.describe()
```

```python
# Isolate rows with null values
null_df = df[df['reamining_contract'].isnull()]
# Display summary stats of rows with null values
null_df.describe()
```

```python
# Isolate rows without null values
not_null_df = df[~df['reamining_contract'].isnull()]
# Display summary stats of rows without null values
not_null_df.describe()
```

```python
# Calculate counts of churned vs. retained
print(df['reamining_contract'].notnull().value_counts())
print()
print(df['reamining_contract'].notnull().value_counts(normalize=True))
```
